###Hypothesis Testing

Hypothesis testing is a statistical method used to make decisions based on sample data.
We test whether the population mean equals 4000.

Null Hypothesis (H₀): μ = 4000

Alternative Hypothesis (H₁): μ ≠ 4000

This is a two-tailed test, as we are checking for any difference from 4000.

```python
# Calculate the t Test Statistic:
sample_mean = 3050
n = 25
populaton_mean = 1000 + 5 * 600
std = 125

# calculating  the t text statistic
# t test is used to determine if there is a significant difference between the means of two groups
# t = (sample_mean - populaton_mean) / (std / sqrt(n))
from math import sqrt
t = (sample_mean - populaton_mean) / (std / (sqrt(n)))

print(f"The test statistic (t) is: {t}")
```

```python
# Determine the Critical Value
from scipy.stats import norm

alpha = 0.05

# the t test is -38.0 so it's greater than the critical value
# For a two-tailed test, we find the critical value at alpha/2
# The critical value is the z-score that corresponds to the tail probabilities
# For a two-tailed test, we use the percent point function (ppf)
critical_value = norm.ppf(1 - alpha / 2)

print(f"The critical value for a two-tailed test with alpha = {alpha} is: \u00B1{critical_value:.2f}")
```

### Conclusion
Based on a sample of 25 restaurants with a mean weekly operating cost of Rs. 3,050, compared to the cost model’s predicted mean of Rs. 4,000, a t-test statistic of –38.0 was calculated.
Since the absolute value of the test statistic (|–38.0| = 38.0) is greater than the critical value (±1.96 at a 5% significance level), the null hypothesis is rejected.
This indicates that there is a statistically significant difference between the sample mean and the model’s predicted mean.
However, because the sample mean is lower than the model’s mean, the results contradict the restaurant owners’ claim that operating costs are higher than predicted.
Instead, the data suggests that actual weekly operating costs are significantly lower than the model’s estimation.
